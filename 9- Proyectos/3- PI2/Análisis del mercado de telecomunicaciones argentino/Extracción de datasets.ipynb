```python
#Creo la ruta de guardado de los Datasets
ruta_carpeta = os.path.abspath('Datasets')
```

```python
def descargar_csv(banco_datos):
    # Descargar archivos CSV
    for diccionario in banco_datos:
        link = f"{url_base}{diccionario['guid']}/data.csv/?auth_key={params['auth_key']}"  #genero el link (ver formato en página web)
        nombre_archivo = f"{diccionario['category_name']}{'_'}{diccionario['title']}.csv"   #genero el nombre del archivo
        ruta_archivo = os.path.join(ruta_carpeta, nombre_archivo)    #le digo en qué carpeta quiero guardarlo

        response = requests.get(link)
        
        if response.status_code == 200:
            with open(ruta_archivo, 'wb') as archivo:
                archivo.write(response.content)
            print(f"Archivo {nombre_archivo} descargado exitosamente.")
        else:
            print(f"No se pudo descargar el archivo {nombre_archivo}. Código de estado: {response.status_code}")
```

```python
descargar_csv(internet_data)
```
